```python
import pandas as pd
import re

# Function to extract horsepower (HP) and RPM from 'Max Power' column
def extract_hp_rpm(power_str):
    if isinstance(power_str, str):
        # Extract HP value (e.g., "118.36bhp@6000rpm" -> 118.36)
        hp_match = re.search(r'(\d+\.?\d*)bhp', power_str)
        hp = float(hp_match.group(1)) if hp_match else None

        # Extract RPM value (e.g., "118.36bhp@6000rpm" -> 6000)
        rpm_match = re.search(r'@(\d+)rpm', power_str)
        rpm = int(rpm_match.group(1)) if rpm_match else None

        return hp, rpm
    return None, None

# Function to calculate max torque
def calculate_torque(hp, rpm):
    if hp and rpm:
        return round((5252 * hp) / rpm, 2)  # Torque in lb-ft, rounded to 2 decimal places
    return None

# Load the Excel file
input_file_path = r'C:/Users/Surenthiran/OneDrive/Desktop/CarDekho/df51.xlsx'
df = pd.read_excel(input_file_path)

# Ensure that the 'Max Power' column is read as string
df['Max Power'] = df['Max Power'].astype(str)

# Apply extraction and torque calculation to each row
df['Horsepower'], df['RPM'] = zip(*df['Max Power'].map(extract_hp_rpm))
df['Max Torque (Calculated)'] = df.apply(lambda row: calculate_torque(row['Horsepower'], row['RPM']), axis=1)

# Convert the calculated torque to Nm (from lb-ft to Nm: 1 lb-ft = 1.3558 Nm)
df['Max Torque (Calculated)'] = df['Max Torque (Calculated)'].apply(lambda x: round(x * 1.3558, 2) if x else None)

# Save the updated DataFrame back to the same Excel file
output_file_path = input_file_path  # Save to the same file
with pd.ExcelWriter(output_file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df.to_excel(writer, index=False, sheet_name='Sheet1')

print(f"Max Torque calculated and saved in {output_file_path}")
```

## **Removing Outliers**

### **Using Z-score Method**

```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your loaded dataset

# Function to calculate and remove outliers using Z-score for all numerical columns
def z_score_analysis(df, threshold=3):
    # Select only numerical columns
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    
    # Calculate Z-scores
    z_scores = np.abs(stats.zscore(df[numerical_cols]))
    
    # Create a boolean mask where True indicates an outlier
    outliers = (z_scores > threshold).any(axis=1)
    
    # Remove outliers
    df_no_outliers = df[~outliers]
    
    # Display number of outliers removed
    print(f"Number of outliers removed: {outliers.sum()}")
    
    return df_no_outliers

# Function to visualize distributions before and after outlier removal
def plot_distribution(df_before, df_after):
    numerical_cols = df_before.select_dtypes(include=[np.number]).columns
    
    for col in numerical_cols:
        # Plot before removing outliers
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        sns.histplot(df_before[col], kde=True)
        plt.title(f'Distribution of {col} (Before)')
        
        # Plot after removing outliers
        plt.subplot(1, 2, 2)
        sns.histplot(df_after[col], kde=True)
        plt.title(f'Distribution of {col} (After)')
        
        plt.tight_layout()
        plt.show()

# Assuming you have loaded your dataset as 'df'

# Apply Z-score analysis to remove outliers
df_cleaned = z_score_analysis(df)

# Visualize the distributions before and after outlier removal
plot_distribution(df, df_cleaned)
```
