```python
import pandas as pd
import re

# Function to extract horsepower (HP) and RPM from 'Max Power' column
def extract_hp_rpm(power_str):
    if isinstance(power_str, str):
        # Extract HP value (e.g., "118.36bhp@6000rpm" -> 118.36)
        hp_match = re.search(r'(\d+\.?\d*)bhp', power_str)
        hp = float(hp_match.group(1)) if hp_match else None

        # Extract RPM value (e.g., "118.36bhp@6000rpm" -> 6000)
        rpm_match = re.search(r'@(\d+)rpm', power_str)
        rpm = int(rpm_match.group(1)) if rpm_match else None

        return hp, rpm
    return None, None

# Function to calculate max torque
def calculate_torque(hp, rpm):
    if hp and rpm:
        return round((5252 * hp) / rpm, 2)  # Torque in lb-ft, rounded to 2 decimal places
    return None

# Load the Excel file
input_file_path = r'C:/Users/Surenthiran/OneDrive/Desktop/CarDekho/df11.xlsx'
df = pd.read_excel(input_file_path)

# Ensure that the 'Max Power' column is read as string
df['Max Power'] = df['Max Power'].astype(str)

# Apply extraction and torque calculation to each row
df['Horsepower'], df['RPM'] = zip(*df['Max Power'].map(extract_hp_rpm))
df['Max Torque (Calculated)'] = df.apply(lambda row: calculate_torque(row['Horsepower'], row['RPM']), axis=1)

# Convert the calculated torque to Nm (from lb-ft to Nm: 1 lb-ft = 1.3558 Nm)
df['Max Torque (Calculated)'] = df['Max Torque (Calculated)'].apply(lambda x: round(x * 1.3558, 2) if x else None)

# Save the updated DataFrame back to the same Excel file
output_file_path = input_file_path  # Save to the same file
with pd.ExcelWriter(output_file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df.to_excel(writer, index=False, sheet_name='Sheet1')

print(f"Max Torque calculated and saved in {output_file_path}")
```

## **Removing Outliers**

### **Using Z-score Method**

```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'df' is your loaded dataset

# Function to calculate and remove outliers using Z-score for all numerical columns
def z_score_analysis(df, threshold=3):
    # Select only numerical columns
    numerical_cols = df.select_dtypes(include=[np.number]).columns
    
    # Calculate Z-scores
    z_scores = np.abs(stats.zscore(df[numerical_cols]))
    
    # Create a boolean mask where True indicates an outlier
    outliers = (z_scores > threshold).any(axis=1)
    
    # Remove outliers
    df_no_outliers = df[~outliers]
    
    # Display number of outliers removed
    print(f"Number of outliers removed: {outliers.sum()}")
    
    return df_no_outliers

# Function to visualize distributions before and after outlier removal
def plot_distribution(df_before, df_after):
    numerical_cols = df_before.select_dtypes(include=[np.number]).columns
    
    for col in numerical_cols:
        # Plot before removing outliers
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        sns.histplot(df_before[col], kde=True)
        plt.title(f'Distribution of {col} (Before)')
        
        # Plot after removing outliers
        plt.subplot(1, 2, 2)
        sns.histplot(df_after[col], kde=True)
        plt.title(f'Distribution of {col} (After)')
        
        plt.tight_layout()
        plt.show()

# Assuming you have loaded your dataset as 'df'

# Apply Z-score analysis to remove outliers
df_cleaned = z_score_analysis(df)

# Visualize the distributions before and after outlier removal
plot_distribution(df, df_cleaned)
```

# **Exploratory Data Analysis (EDA)**

```python
import pandas as pd
import numpy as np

# Assuming 'df' is your loaded dataset

# 1. Overview of the dataset
def dataset_overview(df):
    print("Dataset Shape:", df.shape)
    print("\nData Types:\n", df.dtypes)
    print("\nMissing Values:\n", df.isnull().sum())
    print("\nBasic Statistical Summary:\n", df.describe())  # Summary statistics for numerical columns

# 2. Calculate descriptive statistics manually for a specific column
def descriptive_statistics(df, column):
    mean = df[column].mean()
    median = df[column].median()
    mode = df[column].mode()[0]
    std_dev = df[column].std()
    variance = df[column].var()
    min_val = df[column].min()
    max_val = df[column].max()
    percentiles = df[column].quantile([0.25, 0.50, 0.75])
    
    print(f"\nDescriptive Statistics for {column}:")
    print(f"Mean: {mean}")
    print(f"Median: {median}")
    print(f"Mode: {mode}")
    print(f"Standard Deviation: {std_dev}")
    print(f"Variance: {variance}")
    print(f"Min: {min_val}")
    print(f"Max: {max_val}")
    print(f"25th, 50th, 75th Percentiles:\n{percentiles}")
    
# 3. Descriptive statistics for all numerical columns
def all_numerical_statistics(df):
    print("\nDescriptive Statistics for All Numerical Columns:\n")
    print(df.describe())  # Pandas provides mean, std, min, max, and percentiles

# Assuming you have already loaded your dataset as 'df'
dataset_overview(df)

# Example for analyzing a specific column
# descriptive_statistics(df, 'feature_column')  # Replace 'feature_column' with the column name

# Generate statistics for all numerical columns
all_numerical_statistics(df)
```

# **Feature Selection**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

def read_excel_file(file_path):
    return pd.read_excel(file_path)

def correlation_analysis(df, target_column):
    # Select only numeric columns
    numeric_df = df.select_dtypes(include=['int64', 'float64'])
    corr = numeric_df.corr()[target_column].sort_values(ascending=False)
    plt.figure(figsize=(10, 8))
    sns.barplot(x=corr.index, y=corr.values)
    plt.title('Correlation with Car Price')
    plt.xticks(rotation=90)
    plt.show()
    return corr

def feature_importance(df, target_column):
    # Separate features and target
    X = df.drop(target_column, axis=1)
    y = df[target_column]
    
    # Encode categorical variables
    le = LabelEncoder()
    for column in X.select_dtypes(include=['object']):
        X[column] = le.fit_transform(X[column])
    
    # Train a Random Forest model
    rf = RandomForestRegressor(n_estimators=100, random_state=42)
    rf.fit(X, y)
    
    # Get feature importances
    importances = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
    
    # Plot feature importances
    plt.figure(figsize=(10, 8))
    sns.barplot(x=importances.index, y=importances.values)
    plt.title('Feature Importance for Car Price')
    plt.xticks(rotation=90)
    plt.show()
    
    return importances

# File path
file_path = r"C:/Users/Surenthiran/OneDrive/Desktop/CarDekho/df11.xlsx"

# Read the Excel file
df = read_excel_file(file_path)

# Ensure 'Price of the used car' is numeric
df['Price of the used car'] = pd.to_numeric(df['Price of the used car'], errors='coerce')

# Remove rows with NaN values in the target column
df = df.dropna(subset=['Price of the used car'])

# Perform correlation analysis
print("Correlation Analysis:")
corr_results = correlation_analysis(df, 'Price of the used car')
print(corr_results)

print("\nFeature Importance Analysis:")
importance_results = feature_importance(df, 'Price of the used car')
print(importance_results)

# Combine results
numeric_columns = df.select_dtypes(include=['int64', 'float64']).columns
combined_importance = pd.DataFrame({
    'Correlation': corr_results,
    'Feature Importance': importance_results
})
combined_importance = combined_importance.dropna()
combined_importance['Combined Score'] = combined_importance.mean(axis=1)
combined_importance = combined_importance.sort_values('Combined Score', ascending=False)

print("\nCombined Feature Importance:")
print(combined_importance)

# Plot combined results
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Correlation', y='Feature Importance', data=combined_importance)
for i, txt in enumerate(combined_importance.index):
    plt.annotate(txt, (combined_importance['Correlation'].iloc[i], combined_importance['Feature Importance'].iloc[i]))
plt.title('Correlation vs Feature Importance')
plt.show()
```
