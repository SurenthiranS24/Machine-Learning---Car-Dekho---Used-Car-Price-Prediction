# **Chennai_cars**

```python
import pandas as pd

# Read the Excel file
df = pd.read_excel('C:/Users/Surenthiran/OneDrive/Desktop/CarDekho/chennai_cars.xlsx')

# Display the DataFrame
df
```

```python
import pandas as pd
from io import BytesIO
import ast

# Replace this with the actual path of your Excel file
excel_file_path = "C:/Users/Surenthiran/OneDrive/Desktop/CarDekho/chennai_cars.xlsx"

# Load the Excel file
df = pd.read_excel(excel_file_path, sheet_name=None)

# Extract the relevant sheet
car_data = df['chennai_cars.csv']

# Convert the 'new_car_detail' column from string representation of dict to actual dict
car_data['new_car_detail'] = car_data['new_car_detail'].apply(eval)

# Create a new DataFrame to hold structured data
structured_data = pd.json_normalize(car_data['new_car_detail'])


# Convert the 'new_car_overview' column from string representation of dict to actual dict
car_data['new_car_overview'] = car_data['new_car_overview'].apply(eval)

# Extract relevant fields from the 'new_car_overview'
extracted_data = []
for overview in car_data['new_car_overview']:
    # Assuming 'overview' contains a dictionary with a 'top' key containing a list of dicts with 'key' and 'value'
    details = {item['key']: item['value'] for item in overview['top']}
    extracted_data.append(details)

# Create a new DataFrame from the extracted data
overview_df = pd.DataFrame(extracted_data)

df = pd.read_excel(excel_file_path, sheet_name=0)

# Extract the 'new_car_specs' column
specs_data = df['new_car_specs']

# Initialize a list to hold the extracted values
extracted_rows = []

# Process each entry in the specs data
for entry in specs_data:
    # Convert string representation of dictionary to actual dictionary
    specs_dict = ast.literal_eval(entry)
    
    # Initialize a dictionary to hold the row data
    row_data = {}
    
    # Extract relevant values from the nested structure
    for spec in specs_dict.get('data', []):
        for item in spec.get('list', []):
            if item['key'] in ['Length', 'Width', 'Height', 'Wheel Base', 'Gear Box', 'Top Speed', 'Alloy Wheel Size','Mileage']:
                row_data[item['key']] = item['value']
    
    # Append the row data to the list
    extracted_rows.append(row_data)

# Create a new DataFrame with the extracted data
extracted_data = pd.DataFrame(extracted_rows)

# Initialize a list to hold extracted mileage values
mileage_data = []

# Loop through the 'new_car_specs' column to extract mileage
for entry in df['new_car_specs']:
    # Convert string representation of dictionary to actual dictionary
    specs_dict = ast.literal_eval(entry)
    
    # Initialize a variable to store the mileage for this car
    mileage = None
    
    # Search for mileage data within the 'top' section of the specs
    for item in specs_dict.get('top', []):
        if item.get('key') == 'Mileage':
            mileage = item.get('value')
            break
    
    # Append the extracted mileage data to the list
    mileage_data.append(mileage)

# Create a new DataFrame to store the mileage data
mileage_df = pd.DataFrame({'Mileage': mileage_data})

# Define the features to extract
features_to_extract = [
    "Power Steering", "Power Windows Front", "Air Conditioner", "Heater",
    "Adjustable Head Lights", "Fog Lights Front", "Anti Lock Braking System",
    "Brake Assist", "Radio", "Low Fuel Warning Light", "Accessory Power Outlet",
    "Digital Odometer", "Manually Adjustable Exterior Rear View Mirror",
    "Child Safety Locks"
]

# Create a new DataFrame for the extracted features
extracted_features_df = pd.DataFrame(features_to_extract, columns=["Feature"])

# Convert the 'new_car_specs' column from string representation of dict to actual dict
car_data['new_car_specs'] = car_data['new_car_specs'].apply(eval)

# Extract the specifications and their values
specifications = []
for specs in car_data['new_car_specs']:
    spec_dict = {item['key']: item['value'] for item in specs['data'][0]['list']}
    specifications.append(spec_dict)

# Create a new DataFrame from the specifications
specs_df = pd.DataFrame(specifications)

# Extract car links
car_links = car_data['car_links'].tolist()

# Create a new DataFrame for the car links
car_links_df = pd.DataFrame(car_links, columns=["Car Links"])

# Append all the DataFrames column-wise
combined_df = pd.concat([structured_data, overview_df, extracted_features_df, specs_df, car_links_df,extracted_data,mileage_df], axis=1)

# Save the combined DataFrame to a new Excel file
output_file = "df1_chennai_cars_data.xlsx"
combined_df.to_excel(output_file, index=False)


print(f"Data Extracted and saved to {output_file}")
```

```python
import pandas as pd
from io import BytesIO
import ast

# Replace this with the actual path of your Excel file
excel_file_path = "C:/Users/Surenthiran/OneDrive/Desktop/CarDekho/kolkata_cars.xlsx"

# Load the Excel file
df = pd.read_excel(excel_file_path, sheet_name=None)

# Extract the relevant sheet
car_data = df['kolkata_cars.csv']

# Convert the 'new_car_detail' column from string representation of dict to actual dict
car_data['new_car_detail'] = car_data['new_car_detail'].apply(eval)

# Create a new DataFrame to hold structured data
structured_data = pd.json_normalize(car_data['new_car_detail'])


# Convert the 'new_car_overview' column from string representation of dict to actual dict
car_data['new_car_overview'] = car_data['new_car_overview'].apply(eval)

# Extract relevant fields from the 'new_car_overview'
extracted_data = []
for overview in car_data['new_car_overview']:
    # Assuming 'overview' contains a dictionary with a 'top' key containing a list of dicts with 'key' and 'value'
    details = {item['key']: item['value'] for item in overview['top']}
    extracted_data.append(details)

# Create a new DataFrame from the extracted data
overview_df = pd.DataFrame(extracted_data)

df = pd.read_excel(excel_file_path, sheet_name=0)

# Extract the 'new_car_specs' column
specs_data = df['new_car_specs']

# Initialize a list to hold the extracted values
extracted_rows = []

# Process each entry in the specs data
for entry in specs_data:
    # Convert string representation of dictionary to actual dictionary
    specs_dict = ast.literal_eval(entry)
    
    # Initialize a dictionary to hold the row data
    row_data = {}
    
    # Extract relevant values from the nested structure
    for spec in specs_dict.get('data', []):
        for item in spec.get('list', []):
            if item['key'] in ['Length', 'Width', 'Height', 'Wheel Base', 'Gear Box', 'Top Speed', 'Alloy Wheel Size','Mileage']:
                row_data[item['key']] = item['value']
    
    # Append the row data to the list
    extracted_rows.append(row_data)

# Create a new DataFrame with the extracted data
extracted_data = pd.DataFrame(extracted_rows)

# Initialize a list to hold extracted mileage values
mileage_data = []

# Loop through the 'new_car_specs' column to extract mileage
for entry in df['new_car_specs']:
    # Convert string representation of dictionary to actual dictionary
    specs_dict = ast.literal_eval(entry)
    
    # Initialize a variable to store the mileage for this car
    mileage = None
    
    # Search for mileage data within the 'top' section of the specs
    for item in specs_dict.get('top', []):
        if item.get('key') == 'Mileage':
            mileage = item.get('value')
            break
    
    # Append the extracted mileage data to the list
    mileage_data.append(mileage)

# Create a new DataFrame to store the mileage data
mileage_df = pd.DataFrame({'Mileage': mileage_data})

# Define the features to extract
features_to_extract = [
    "Power Steering", "Power Windows Front", "Air Conditioner", "Heater",
    "Adjustable Head Lights", "Fog Lights Front", "Anti Lock Braking System",
    "Brake Assist", "Radio", "Low Fuel Warning Light", "Accessory Power Outlet",
    "Digital Odometer", "Manually Adjustable Exterior Rear View Mirror",
    "Child Safety Locks"
]

# Create a new DataFrame for the extracted features
extracted_features_df = pd.DataFrame(features_to_extract, columns=["Feature"])

# Convert the 'new_car_specs' column from string representation of dict to actual dict
car_data['new_car_specs'] = car_data['new_car_specs'].apply(eval)

# Extract the specifications and their values
specifications = []
for specs in car_data['new_car_specs']:
    spec_dict = {item['key']: item['value'] for item in specs['data'][0]['list']}
    specifications.append(spec_dict)

# Create a new DataFrame from the specifications
specs_df = pd.DataFrame(specifications)

# Extract car links
car_links = car_data['car_links'].tolist()

# Create a new DataFrame for the car links
car_links_df = pd.DataFrame(car_links, columns=["Car Links"])

# Append all the DataFrames column-wise
combined_df = pd.concat([structured_data, overview_df, extracted_features_df, specs_df, car_links_df,extracted_data,mileage_df], axis=1)

# Save the combined DataFrame to a new Excel file
output_file = "df6_kolkata_cars_data.xlsx"
combined_df.to_excel(output_file, index=False)


print(f"Data Extracted and saved to {output_file}")
```

```python
import pandas as pd
import re

# Function to extract horsepower (HP) and RPM from 'Max Power' column
def extract_hp_rpm(power_str):
    if isinstance(power_str, str):
        # Extract HP value (e.g., "118.36bhp@6000rpm" -> 118.36)
        hp_match = re.search(r'(\d+\.?\d*)bhp', power_str)
        hp = float(hp_match.group(1)) if hp_match else None

        # Extract RPM value (e.g., "118.36bhp@6000rpm" -> 6000)
        rpm_match = re.search(r'@(\d+)rpm', power_str)
        rpm = int(rpm_match.group(1)) if rpm_match else None

        return hp, rpm
    return None, None

# Function to calculate max torque
def calculate_torque(hp, rpm):
    if hp and rpm:
        return round((5252 * hp) / rpm, 2)  # Torque in lb-ft, rounded to 2 decimal places
    return None

# Load the Excel file
input_file_path = r'C:/Users/Surenthiran/OneDrive/Desktop/CarDekho/df11.xlsx'
df = pd.read_excel(input_file_path)

# Ensure that the 'Max Power' column is read as string
df['Max Power'] = df['Max Power'].astype(str)

# Apply extraction and torque calculation to each row
df['Horsepower'], df['RPM'] = zip(*df['Max Power'].map(extract_hp_rpm))
df['Max Torque (Calculated)'] = df.apply(lambda row: calculate_torque(row['Horsepower'], row['RPM']), axis=1)

# Convert the calculated torque to Nm (from lb-ft to Nm: 1 lb-ft = 1.3558 Nm)
df['Max Torque (Calculated)'] = df['Max Torque (Calculated)'].apply(lambda x: round(x * 1.3558, 2) if x else None)

# Save the updated DataFrame back to the same Excel file
output_file_path = input_file_path  # Save to the same file
with pd.ExcelWriter(output_file_path, engine='openpyxl', mode='a', if_sheet_exists='replace') as writer:
    df.to_excel(writer, index=False, sheet_name='Sheet1')

print(f"Max Torque calculated and saved in {output_file_path}")
```
